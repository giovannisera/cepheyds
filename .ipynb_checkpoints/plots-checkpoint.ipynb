$$M_v=c_1+c_2 logP$$

$$c1=-1.2055611181547214, \ c2=-2.9000855979736215$$

```python
filename1 = "tab_ceph.txt"
filename2 = "periodi_ceph.txt"
data1 = np.genfromtxt(filename1)
data1 = np.transpose(data1)
data2 = np.genfromtxt(filename2)
data2 = np.transpose(data2)
```

```python
#cefeidi già analizzate
P = data1[0]
M_v = data1[1]
#cefeidi analizzate da me
P2 = data2[0]
P2_err = data2[1]
P2_err = (1 /P2)*P2_err
M_v2 = data2[4]
M_v2_err = data2[5]
c1 = -1.2055611181547214
c2 = -2.9000855979736215
max1, max2 = np.max(P), np.max(P2)
maximum = np.array([max1, max2])
max = np.max(maximum)
min1, min2 = np.min(P), np.min(P2)
minimum = np.array([min1, min2])
min = np.min(minimum)
x = np.arange(np.log10(min), np.log10(max),0.01)
y = c1 + c2*x
```

```python
a = plt.figure("M_V vs LOG_{10}P", figsize=(18,7))
gs = gridspec.GridSpec(1,2)
gs.update(wspace=0.1)
ax1 = plt.subplot(gs[0])
ax2 = plt.subplot(gs[1])

ax1.scatter(np.log10(P), M_v, s=10, marker='s', color = 'r', edgecolors='r', label='ceph_catalog', zorder=2)
ax1.scatter(np.log10(P2), M_v2, s=10, marker='p',color = "grey", edgecolors='black', label='Sera G. 2023', zorder=2)
ax1.errorbar(np.log10(P2), M_v2, xerr=(P2_err, P2_err), yerr=(M_v2_err,M_v2_err),
                 fmt='none', ecolor='black', alpha=1, elinewidth=1, capsize=3, zorder=1, label='Errori')
ax1.plot(x , y, c='b', label='Best-Fit', zorder=0)
ax1.set_xlabel('log$_{10}$P', fontsize = 12)
ax1.set_ylabel('M$_\mathrm{v}$', fontsize = 12, rotation = 0, labelpad=16)
ax1.set_ylim(np.max(M_v) + 0.5, np.min(M_v) - 0.2)
ax1.set_yticks(np.arange(-1.5, -6, -0.5))
ax1.set_xticks(np.arange(0.1, 1.6, 0.1))
ax1.legend(fontsize="small")

ax2.scatter(np.log10(P), M_v, s=10, marker='s', color = 'r', edgecolors='r', label='ceph_catalog', zorder=2)
ax2.scatter(np.log10(P2), M_v2, s=10, marker='p',color = "grey", edgecolors='black', label='Sera G. 2023', zorder=2)
ax2.errorbar(np.log10(P2), M_v2, xerr=(P2_err, P2_err), yerr=(M_v2_err,M_v2_err),
                 fmt='none', ecolor='black', alpha=1, elinewidth=1, capsize=3, zorder=1, label='Errori')
ax2.plot(x , y, c='b', label='Best-Fit', zorder=0)
ax2.set_xlabel('log$_{10}$P', fontsize = 12)
ax2.set_xlim(0.85, 1.5)
ax2.set_ylim(-3.5, - 5.5)
ax2.yaxis.tick_right()
ax2.yaxis.set_label_position("right")
#ax1.set_yticks(np.arange(-1.5, -6, -0.5))
#ax1.set_xticks(np.arange(0.1, 1.6, 0.1))
#ax2.legend(fontsize="x-small")

os.chdir(dir_plot)
plt.savefig("1_per-mag.png", dpi=500)
os.chdir(dir_home)
```

$$CURVE \ \ DI \ \ LUCE$$

```python
filename_a = "tab_ceph4536a.txt" 
filename_b = "tab_ceph4536b.txt"  
filename_c = "tab_ceph4536c.txt"
filename_spline_a = "spline_ceph4536a.txt"  
filename_spline_b = "spline_ceph4536b.txt" 
filename_spline_c = "spline_ceph4536c.txt" 
data_a = np.genfromtxt (filename_a)
data_a = np.transpose (data_a)
data_b = np.genfromtxt (filename_b)
data_b = np.transpose (data_b)
data_c = np.genfromtxt (filename_c)
data_c = np.transpose (data_c)
spline_a = np.genfromtxt (filename_spline_a)
spline_a = np.transpose (spline_a)
spline_b = np.genfromtxt (filename_spline_b)
spline_b = np.transpose (spline_b)
spline_c = np.genfromtxt (filename_spline_c)
spline_c = np.transpose (spline_c)
```

```python
#NGC4536a
time_a = data_a[0]
m_app_a = data_a[1]
err_a = data_a[2]
x_a = spline_a[0]
y_a = spline_a[1]
#NGC4536b
time_b = data_b[0]
m_app_b = data_b[1]
err_b = data_b[2]
x_b = spline_b[0]
y_b = spline_b[1]
#NGC4536c
time_c = data_c[0]
m_app_c = data_c[1]
err_c = data_c[2]
x_c = spline_c[0]
y_c = spline_c[1]
```

```python
a = plt.figure("Curve di Luce", figsize=(21,7))
gs = gridspec.GridSpec(1,3)
gs.update(wspace=0.2)
ax1 = plt.subplot(gs[0])
#ax1.set_title("NGC4536a")
ax2 = plt.subplot(gs[1])
#ax2.set_title("NGC4536b")
ax3 = plt.subplot(gs[2])
#ax3.set_title("NGC4536c")

ax1.scatter(time_a ,m_app_a , s=0.8, marker='s', edgecolors='b', label='ceph_NGC4536a', zorder=2)
ax1.errorbar(time_a, m_app_a, yerr=(err_a,err_a), fmt='none', ecolor='red', alpha=1, elinewidth=1, capsize=3, zorder=1)
ax1.plot(x_a, y_a, c='grey', zorder=0)
ax1.set_ylim(np.max(m_app_a) + 0.3, np.min(m_app_a) - 0.2)
ax1.set_xlabel('t (JD)', fontsize = 12)
ax1.set_ylabel('m$_\mathrm{V}$', fontsize = 14, rotation = 0, labelpad=14)
ax1.legend(loc="lower left")

ax2.scatter(time_b, m_app_b, s=0.8, marker='s', edgecolors='b', label='ceph_NGC4536b', zorder=2)
ax2.errorbar (time_b, m_app_b, yerr=(err_b,err_b), fmt='none', ecolor='red', alpha=1, elinewidth=1, capsize=3, zorder=1)
ax2.plot(x_b, y_b, c='grey', zorder=0)
ax2.set_ylim(np.max(m_app_b) + 0.3, np.min(m_app_b) - 0.2)
ax2.set_xlabel('t (JD)', fontsize = 12)
#ax2.set_ylabel('$m_V$', fontsize = 14, rotation = 0, labelpad=14)
ax2.legend(loc="lower left")

ax3.scatter(time_c, m_app_c, s=0.8, marker='s', edgecolors='b', label='ceph_NGC4536c', zorder=2)
ax3.errorbar (time_c, m_app_c, yerr=(err_c,err_c), fmt='none', ecolor='red', alpha=1, elinewidth=1, capsize=3, zorder=1)
ax3.plot(x_c, y_c, c='grey', zorder=0)
ax3.set_ylim(np.max(m_app_c) + 0.3, np.min(m_app_c) - 0.2)
ax3.set_xlabel('t (JD)', fontsize = 12)
#ax3.set_ylabel('m$_\mathrm{V}$', fontsize = 14, rotation = 0, labelpad=14)
ax3.legend(loc="lower left")

os.chdir(dir_plot)
plt.savefig("2_curve_di_luce_4536.png", dpi=500)
os.chdir(dir_home)
```

$$\tilde{\chi}^2$$

```python
filename_chi_a = "chiquadro_ceph4536a.txt" 
filename_chi_b = "chiquadro_ceph4536b.txt" 
filename_chi_c = "chiquadro_ceph4536c.txt"
chi_a = np.genfromtxt (filename_chi_a)
chi_a = np.transpose (chi_a)
chi_b = np.genfromtxt (filename_chi_b)
chi_b = np.transpose (chi_b)
chi_c = np.genfromtxt (filename_chi_c)
chi_c = np.transpose (chi_c)
```

```python
#NGC4536a
period_a = chi_a[0]
chisq_a = chi_a[1]
#NGC4536b
period_b = chi_b[0]
chisq_b = chi_b[1]
#NGC4536c
period_c = chi_c[0]
chisq_c = chi_c[1]
```

```python
a = plt.figure("chi-quadro", figsize=(21,7))
gs = gridspec.GridSpec(1,3)
gs.update(wspace=0.2)

ax1 = plt.subplot(gs[0])
#ax1.set_title("$NGC4536a$")
ax2 = plt.subplot(gs[1])
#ax2.set_title("$NGC4536b$")
ax3 = plt.subplot(gs[2])
#ax3.set_title("$NGC4536c$")

ax1.plot(period_a, chisq_a, c='g', zorder=1, label="NGC4536a") 
ax1.set_xlabel('periodo (JD)', fontsize = 12)
ax1.set_ylabel(r'$\tilde{\chi}^2$', fontsize = 14, rotation = 0, labelpad=14)
ax1.text(0, 100, "T=16.80±0.10 Jd", fontsize=14)
ax1.legend()

ax2.plot(period_b, chisq_b, c='g', zorder=1, label="NGC4536b")
ax2.set_xlabel('periodo (JD)', fontsize = 12)
#ax2.set_ylabel(r'$\tilde{\chi}^2$', fontsize = 14, rotation = 0, labelpad=14)
ax2.text(0, 50, "T=12.24±0.05 Jd", fontsize=14)
ax2.legend()

ax3.plot(period_c, chisq_c, c='g', zorder=1, label="NGC4536c")
ax3.set_xlabel('periodo (JD)', fontsize = 12)
#ax3.set_ylabel(r'$\tilde{\chi}^2$', fontsize = 14, rotation = 0, labelpad=14)
ax3.text(0, 320, "T=18.60±0.04 Jd", fontsize=14)
ax3.legend()

os.chdir(dir_plot)
plt.savefig("3_chi-quadro_4536.png", dpi=500)
os.chdir(dir_home)
```

$$COSMOLOGIA$$

$$LEGGE \ \ DI \ \ HUBBLE$$

```python
filename = "hubble.txt"
data = np.genfromtxt("hubble.txt")
data = np.transpose(data)
```

```python
#distanza con errore
d = data[0]
de = data[1]
#velocità di recessione con errore
v = data[2]
ve = data[3]
#array contentente le costanti di Hubble trvate
H0 = data[4]
# array delle distanze per tracciare la legge di Hubble
x = np.arange(0, 26, 0.01)
```

```python
a = plt.figure("Hubble's law", figsize=(7,5))
ax = a.add_subplot(111)
ax.scatter(d, v, s=10, marker='s', edgecolors='k', zorder=2)
ax.errorbar(d, v, xerr=(de, de), yerr=(ve, ve), fmt='none', ecolor='black', alpha=1, elinewidth=1, capsize=3, zorder=1)
for i in range(10):
    plt.plot(x , x*H0[i], c='grey', linestyle="--" ,zorder=0)
ax.plot(x, x*76.54 , c='red', label='media', zorder = 0)
ax.set_xlim(0, 25)
ax.set_ylim(-250, 2250)
ax.set_xlabel("d (Mpc)")
ax.set_ylabel("v$_{\mathrm{rec}}$ (Mpc sec$^{-1}$)")
ax.tick_params(bottom=True, top=True, left=True, right=True)
ax.tick_params(labelbottom=True, labeltop=False, labelleft=True, labelright=False)
ax.tick_params(axis="x", direction="in", length=7, width=1)
ax.tick_params(axis="y", direction="in", length=7, width=1)

plt.text(18.5, 2050, "100", fontsize = 15) #max
plt.text(22, 1300, "63", fontsize = 15) #min
plt.text(23.5, 1800, "77", fontsize = 15) #media

ax.legend()

os.chdir(dir_plot)
plt.savefig("4_hubble.png", dpi=500)
os.chdir(dir_home)
```

$$PARAMETRI \ \ DI \ \ DENSITA', \ \ H_0=76.54\pm1.22 \ \frac{km}{s \ Mpc}$$

```python
filename_1sigma = "1sigma.txt"
filename_2sigma = "2sigma.txt"
filename_3sigma = "3sigma.txt"
###  1sigma-2sigma-3sigma
sigma1 = np.genfromtxt(filename_1sigma)
sigma1 = np.transpose(sigma1)
sigma2 = np.genfromtxt(filename_2sigma)
sigma2 = np.transpose(sigma2)
sigma3 = np.genfromtxt(filename_3sigma)
sigma3 = np.transpose(sigma3)
```

```python
a = plt.figure("Cosmologia", figsize = (6,6), dpi = 120)
#a.suptitle("Età dell'Universo" + "\n" + "H$_0$=76.54$\pm$1.22, t$_0$=13.82$\pm$0.14 Mld yy (Planck 2018)", fontsize = 10)
ax = a.add_subplot(111)
plt.xlabel('$\Omega_{M}$', fontsize = 14)
plt.ylabel(r'$\Omega_{\Lambda}$', fontsize = 14, rotation = 0)
plt.ylim(-0.2,1)
ax.yaxis.tick_right()
ax.yaxis.set_label_position("right")
plt.xlim(0,1.2)
ax.tick_params(bottom=True, top=True, left=True, right=True)
ax.tick_params(labelbottom=True, labeltop=False, labelleft=False, labelright=True)
ax.tick_params(axis="x", direction="in", length=7, width=1)
ax.tick_params(axis="y", direction="in", length=7, width=1)
###
x1 = np.arange(0, 1.21, 0.1)
#UNIVERSO PIATTO (OM + OL = 1)
ax.plot(x1, 1 - x1, color='black', linestyle='dashed', 
     linewidth=0.7, alpha=0.6, zorder=3)
#1 SIGMA
ax.scatter(sigma1[0], sigma1[1], c='yellow', marker='o', 
      s=0.1, alpha=1, label = "1$\sigma$", zorder = 2)
#2 SIGMA
ax.scatter(sigma2[0], sigma2[1], color='red', marker='o', 
     s=0.1, alpha=1, label = "2$\sigma$", zorder = 1)
#3 SIGMA
ax.scatter(sigma3[0], sigma3[1], color='blue', marker='o', 
     s=0.11, alpha=1, label = "3$\sigma$", zorder = 0)
#REMARKS
ax.scatter(x = 0.200, y = 0.800, marker = 's', edgecolors = 'black', s=6, alpha = 1, zorder = 4, label = '[0.200, 0.800]')
#cambiarlo con la coppia di parametri trovati con il mio H_0
plt.text(0.74, 0.14, "piatto", rotation = -45, fontsize = 15)
plt.text(0.4, 0.1, "aperto", fontsize = 15)
plt.text(0.8, 0.5, "chiuso", fontsize = 15)

plt.legend(fontsize = "x-small")

os.chdir(dir_plot)
plt.savefig("5_parameters.png", dpi=500)
os.chdir(dir_home)
```
